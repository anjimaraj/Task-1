```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw  
```

```python
for i in ['Age','SibSp','Fare']:
    lw,uw=wisker(df[i])
    df1[i]=np.where(df1[i]<lw,lw,df1[i])
    df1[i]=np.where(df1[i]>uw,uw,df1[i])
for i in ['Age','SibSp','Fare']:
    sns.boxplot(df1[i])
    plt.show()
```

```python
std=StandardScaler()
output=df1['Survived']
inputs=df1.drop(['Survived'],axis=1)
output
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,output,test_size=0.3,random_state=0)
print(inputs.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
models=[[SVC(),'Support vector'],
        [LogisticRegression(C=10),'LogisticRegression'],
        [RandomForestClassifier(n_estimators=20),'RandomForestClassifier'],
        [DecisionTreeClassifier(max_depth=7),'DecisionTreeClassifier'],
        [KNeighborsClassifier(n_neighbors=7),'KNeighborsClassifier']]
```

```python
for i in models:
    name=i[1]
    model=i[0]
    print(name)
    model.fit(X_train,y_train)
    
    y_pred=model.predict(X_test)
    accuracy=accuracy_score(y_test,y_pred)
    precision=precision_score(y_test,y_pred)
    recall=recall_score(y_test,y_pred)
    
    f1=f1_score(y_test,y_pred)
    print("Accuracy=",accuracy)
    print("Precision=",precision)
    print("Recall=",recall)
    print("F1 score=",f1)
    
    conf_matrix=confusion_matrix(y_test,y_pred)
    print("Confusion matrix=\n",conf_matrix)
    
    sns.heatmap(conf_matrix,annot=True,fmt='d',cmap='Blues')
    plt.ylabel('Actual')
    plt.xlabel('Prediction')
    plt.show()
    print("\n")
```
